```python
# import pandas library
import pandas as pd
```

```python
# read the dataset
df = pd.read_excel(r"C:\Users\yoges\OneDrive\Desktop\Analytics project\Python_Diwali_Sales_Analysis-main\Customer_Call_List\Customer Call List.xlsx")
df
```

```python
# remove any duplicate rows
df = df.drop_duplicates()
df
```

```python
# drop unnecessary columns
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# remove unnecessary characters from names
df['Last_Name'] = df['Last_Name'].str.strip('/..._')
df
```

```python
# format phone numbers in a uniform style
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
# split the address into street name, state, and zip code
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',2,expand=True)
df
```

```python
# use uniform text in columns 'Paying Customer' and 'Do_Not_Contact'
df['Paying Customer'] = df['Paying Customer'].apply(lambda x: (x == 'Y' and x + 'es') or (x == 'N' and x + 'o') or x)
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: (x == 'Y' and x + 'es') or (x == 'N' and x + 'o') or x)
df
```

```python
# replace N/a and NaN with blank values
df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
# remove customers who do not wish to be contacted
df = df[df['Do_Not_Contact']!='Yes']
df
```

```python
# remove customers whose phone numbers are not available
df = df[df['Phone_Number']!='']
df
```

```python
# reset the index (helps in identifying row index)
df = df.reset_index(drop=True)
df
```
